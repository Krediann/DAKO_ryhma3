#### Precision-recall curve

Precision-recall curve is better for imbalanced datasets. In Titanic dataset the survival rate is around 38 percent so the imbalance is mild. Thus, it's probably enough to use ROC curve to summarize the trade-off between the true positive rate and false positive rate for a predictive model. But we'll do the precision-recall curve here too.

```python
y_scores_svm = cross_val_predict(svm_clf, X_train, y_train, cv=7,
                             method="decision_function")
precisions, recalls, thresholds = precision_recall_curve(y_train, y_scores_svm)
```

Because the Decision tree is designed only to give a class prediction (it's a descrete model) it only yields single confuction matrix. This means that the ROC curve and Precision-recall curve cannot be plotted similar with the SVM model. <br>
<br>
As far as we have understood, the way to circumewent this, is to use "predict_proba method that calculates the proportion of positive class versus all other data in a leaf where the prediction is end up with. The course book says that the method "returns an array containing a row per instance and a column per class, each containing the probability that the given
instance belongs to the given class."
<br><br>
Sources: <br>
https://blog.zenggyu.com/en/post/2019-03-03/how-to-plot-roc-and-precision-recall-curves/ <br>
https://machinelearningmastery.com/threshold-moving-for-imbalanced-classification/ <br>
https://machinelearningmastery.com/roc-curves-and-precision-recall-curves-for-classification-in-python/ <br>
https://developers.google.com/machine-learning/data-prep/construct/sampling-splitting/imbalanced-data#:~:text=A%20classification%20data%20set%20with,smaller%20proportion%20are%20minority%20classes.

```python
y_proba_dectree = cross_val_predict(decision_tree, X_train, y_train, cv=7,
                             method="predict_proba")
#instead of score we give the class probabilities for the curve (if not given we get an error:
# "y should be a 1d array, got an array of shape (891, 2) instead", when computing the precisions, recalls and tresholds)
y_proba_dectree = y_proba_dectree [:, 1]
```

```python
tree_prec, tree_rec, tree_tresh = precision_recall_curve(y_train, y_proba_dectree)
```

```python
def plot_precision_vs_recall(precisions, recalls, label=None):
    plt.plot(recalls, precisions, linewidth=2, label=label)
    plt.xlabel("Recall", fontsize=16)
    plt.ylabel("Precision", fontsize=16)
    plt.axis([0, 1, 0, 1])
    plt.grid(True)

plt.figure(figsize=(8, 6))
plot_precision_vs_recall(precisions, recalls, label="SVM")
plot_precision_vs_recall(tree_prec, tree_rec, label="Decision Tree")
#plt.plot([recall_90_precision, recall_90_precision], [0., 0.9], "r:")
#plt.plot([0.0, recall_90_precision], [0.9, 0.9], "r:")
#plt.plot([recall_90_precision], [0.9], "ro")
plt.grid(True)
plt.legend(loc="lower right", fontsize=16)
save_fig("precision_vs_recall_plot")
plt.show()
```

##### ROC AUC curve

To compare classifiers we can measure the area under the curve (AUC). A perfect classifier will have a ROC AUC equal to 1, whereas a purely random classifier will have a ROC AUC equal to 0.5

1. Smaller values on the x-axis of the plot indicate lower false positives and higher true negatives, i.e., the model is predicting the negatives correctly.
2. Larger values on the y-axis of the plot indicate higher true positives and lower false negatives, i.e., the model is preicting the positives correctly.

```python
# the ROC curve plots sensitivity (recall) versus 1 – specificity
#ROC curve for svm
fpr, tpr, thresholds = roc_curve(y_train, y_scores_svm)
```

```python
svm_auc = roc_auc_score(y_train, y_scores_svm)
```

```python
# ROC curve for Decision tree
#y_scores_dectree = y_proba_dectree[:, 1] # score = probability of a positive class
fpr_dectree, tpr_dectree, thresholds_dectree = roc_curve(y_train,y_proba_dectree)
```

```python
dectree_auc = roc_auc_score(y_train, y_proba_dectree)
```

```python
def plot_roc_curve(fpr, tpr, label=None):
    plt.plot(fpr, tpr, linewidth=2, label=label)
    plt.plot([0, 1], [0, 1], 'k--') # dashed diagonal
    plt.axis([0, 1, 0, 1])                                    # Not shown in the book
    plt.xlabel('False Positive Rate (Fall-Out)', fontsize=16) # Not shown
    plt.ylabel('True Positive Rate (Recall)', fontsize=16)    # Not shown
    plt.grid(True)                                            # Not shown

plt.figure(figsize=(8, 6))                                    # Not shown
plot_roc_curve(fpr, tpr, label="SVM")
plot_roc_curve(fpr_dectree, tpr_dectree, label="Decision Tree")
plt.grid(True)
plt.legend(loc="lower right", fontsize=16)
save_fig("roc_curve_plot")                                    # Not shown
plt.show()
```

gini or enthropy
splitter: string, optional (default=”best”) it evaluate all splits using the criterion before splitting; random splitter uses a random uniform function with min_feature_value, max_feature_value and random_state as inputs.
max_depth: int or None, optional (default=None)
min_samples_split: int, float, optional (default=2)
the ideal min_samples_split values tend to be between 1 to 40 for the CART algorithm which is the algorithm implemented in scikit-learn.
 the ideal min_samples_leaf values tend to be between 1 to 20 for the CART algorithm

 max_features: int, float, string or None, optional (default=None)
 random_state: int, RandomState instance or None, optional (default=None) not a hyperparameter!

```python
param_grid = [
    {"criterion": ["gini", "entropy"], 
     "max_depth": [2, 4, 6, 8, 10, 12, 14],
     "min_samples_split": [2, 3, 4],
     "min_samples_leaf": [5, 15, 25, 35],
     "max_leaf_nodes": list(range(2, 50)),
     "max_features": [None, "auto", "sqrt", "log2"],
     "class_weight": [None, "balanced"]
    }
]
grid_search = GridSearchCV(DecisionTreeClassifier(random_state=42), 
                           param_grid, verbose=1, cv=5)
```
